# Profession
## Artists makeup the largest portion of their business.

```python
import pandas as pd

data = pd.read_csv('Customers.csv')

counts = data['Profession'].value_counts()
print(counts)


profession = df.Profession.value_counts()


names = profession.index

fig = px.pie(values=profession, names=names, color=names)


fig.update_layout(title_text='Profession')
fig.update_traces(pull=[0.1, 0, 0, 0])

fig.show()
```

# Spending Score (1-100) by Profession
## The store created a spending score based on the frequency of purchases by customers over a year. Artists and those in Entertainment have the highest Spending Score. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('Customers.csv')

sns.set(style="whitegrid")
plt.figure(figsize=(8,6))
sns.barplot(x='Spending Score (1-100)', y='Profession', data=data)
plt.title('Spending Score by Profession')
plt.xlabel('Spending Score')
plt.ylabel('Profession')
plt.show()
```

# Customer Family Size

# Distribution of all customer data by profession and spending score. The Artist column shows they have larger family sizes than other professions. 

```python
import plotly.express as px
import pandas as pd

data = pd.read_csv('Customers.csv')
data = data.dropna()

fig = px.density_heatmap(data_frame=data, x='Annual Income ($)', y='Spending Score (1-100)', z='Family Size', facet_col='Profession', color_continuous_scale='YlOrRd', nbinsx=10, nbinsy=10)


fig.show()
```

# Family size is represented by the target size. 

```python
import plotly.express as px
import pandas as pd

data = pd.read_csv('Customers.csv')

fig = px.scatter(data_frame=data, x='Annual Income ($)', y='Spending Score (1-100)', size='Family Size')

#fig.update_traces(line=dict(dash='dash'))

fig.show()
```

# An overview of customers by spending score, profession and gender.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('Customers.csv')

sorted_df = df.sort_values(by=['Profession', 'Annual Income ($)'], ascending=[False, True])

fig = px.scatter(sorted_df, x='Profession', y='Spending Score (1-100)', color='Gender')

fig.show()
```
